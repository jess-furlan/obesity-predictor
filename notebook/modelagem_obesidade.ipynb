Entender a distribuição das variáveis, identificar valores nulo, outliers e correlações.



---


**1. Verificação dos Tipos de Dados**

É preciso garantir que cada coluna esteja representada com o tipo correto (float, int, object, category, etc.)

**Por que é essencial:** Algoritmos de Machine Learning exigem entrada numérica. Variáveis categóricas precisam ser codificadas corretamente.
Um campo numérico mal carregado como object pode impedir transformações como normalização, gerar erros ou distorcer estatísticas.

**Impacto no modelo:** evita falhas no pipeline e garante que pré-processamentos como normalização, encoding e modelagem sejam realizados corretamente.



---


**2. Verificação de Valores Nulos**

Identificar e tratar dados ausentes, o que pode indicar problemas na coleta, entrada ou carregamento dos dados.

**Por que é essencial:** Modelos de Machine Learning geralmente não lidam bem com valores nulos — eles precisam ser imputados (preenchidos) ou removidos.
Valores ausentes em variáveis críticas podem introduzir viés se não tratados com cuidado.

**Impacto no modelo:** evita erro de execução e melhora a robustez e generalização do modelo.



---


**3. Estatísticas Descritivas**

Observar a média, desvio padrão, mínimo, máximo, frequências — para entender a distribuição, dispersão e presença de outliers.

**Por que é essencial:** Permite identificar escalas diferentes entre variáveis, o que exige normalização.
Valores extremos ou inconsistentes podem indicar outliers ou erros de digitação, que afetam negativamente a modelagem.

**Impacto no modelo:** melhora a convergência e performance de modelos sensíveis à escala e evita aprendizado incorreto.



---

Entender visualmente como as variáveis se relacionam com a variável alvo (Obesity) e entre si.

# Pré-processamento dos Dados e Feature Engineering



---


**1. Codificação dos Dados Categóricos (Label Encoding)**

transformar categorias de texto (Gender, FAVC, etc.) em números, pois a maioria dos algoritmos de ML não trabalha com strings.

**Por que LabelEncoder:** Simples e direto para modelos baseados em árvore (como Random Forest, XGBoost), que não são sensíveis à ordem dos números.

Permitirá salvar os encoders para uso futuro no deploy (no Streamlit).


---


**2. Criação da Feature Derivada IMC**

Formula é peso dividido pelo quadrado da altura.

**Por que é essencial:** O IMC é uma variável composta usada globalmente por profissionais da saúde para classificar níveis de obesidade.
Ajuda o modelo a entender melhor a relação peso-altura de forma mais significativa do que usar essas variáveis isoladamente.




---


**3. Padronização com StandardScaler**

todos os atributos contínuos foram normalizados para média 0 e desvio 1.

**Por que é essencial:** Algoritmos como SVM, KNN e redes neurais são sensíveis à escala.
Garante que nenhuma variável domine o modelo apenas por sua escala maior (ex: Weight vs FAF).

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Cópia do dataframe para preservar original
df_processed = df.copy()

# --------------------------
# 1. Codificação dos Dados Categóricos

# Lista de colunas categóricas
cat_cols = ['Gender', 'family_history', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS', 'Obesity']

# Dicionário para armazenar os codificadores (útil para usar no deploy depois)
label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    df_processed[col] = le.fit_transform(df_processed[col])
    label_encoders[col] = le  # guardar o codificador para uso posterior

# --------------------------
# 2. Criação da Feature IMC

df_processed['BMI'] = df_processed['Weight'] / (df_processed['Height'] ** 2)

# --------------------------
# 3. Padronização dos Dados Numéricos

# Selecionar colunas numéricas contínuas
num_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE', 'BMI']

scaler = StandardScaler()
df_processed[num_cols] = scaler.fit_transform(df_processed[num_cols])

# Separar X e y
X = df_processed.drop('Obesity', axis=1)
y = df_processed['Obesity']

# Verificar resultado do pré-processamento
X.head()
```

# Construção do Modelo de Machine Learning
**Algoritmos a serem testados:**

*   Random Forest
*   XGBoost
*   LightGBM
*   SVM

**Avaliação:** accuracy, f1-score, confusion matrix

**Critério de escolha:** Acurácia acima de 75%

Validação cruzada com StratifiedKFold




---


**1. Teste de múltiplos modelos**

Cada algoritmo foi escolhido estrategicamente:

* Random Forest: robusto, boa baseline, resistente a overfitting, lida bem com dados mistos e não exige muita parametrização.
* XGBoost: ótima performance em muitos tipos de problema, com controle fino.
* SVM: simples, útil em alta dimensionalidade.
* LightGBM: rápido e eficaz em datasets grandes, similar ao XGBoost


---

**2. Validação cruzada com StratifiedKFold**

A validação cruzada com StratifiedKFold divide os dados mantendo a proporção das classes da variável alvo, garantindo que todas estejam representadas.
Isso é essencial em problemas de classificação com classes desbalanceadas, garantindo avaliação justa e sem viés.



---


**3. Métrica de Avaliação: Acurácia**

Avaliação dos modelos pela acurácia média e desvio padrão nos folds.
Isso garante que o modelo seja estável e generalizável.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np

# --------------------------
# 1. Definição dos Modelos

models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42)
}

# --------------------------
# 2. Validação Cruzada

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# --------------------------
# 3. Avaliação dos Modelos

results = {}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
    results[name] = {
        "Accuracy Mean": scores.mean(),
        "Accuracy Std": scores.std()
    }

# Organizar os resultados em DataFrame para visualização
results_df = pd.DataFrame(results).T.sort_values(by="Accuracy Mean", ascending=False)
results_df
```

---

**Análise parcial dos resultados**

Esses resultados mostram que o pipeline está funcionando com altíssima performance:

| Modelo            | Acurácia Média | Desvio Padrão |
| ----------------- | -------------- | ------------- |
| **Random Forest** | **99.19%**     | ±0.28%        |
| XGBoost           | 98.63%         | ±0.48%        |
| LightGBM          | 98.63%         | ±0.51%        |
| SVM               | 93.79%         | ±1.07%        |

* Todos os modelos superaram o critério mínimo de 75% de acurácia proposto no desafio.
* O Random Forest foi o modelo com melhor performance média e menor variação entre os folds, sendo o mais estável.
* XGBoost e LightGBM também são muito bons, e geralmente superiores em datasets maiores ou mais ruidosos.
* O SVM, embora abaixo dos outros, ainda teve bom desempenho, mas com maior instabilidade.

Um modelo com acurácia alta e variação baixa tende a se comportar bem no mundo real, especialmente em ambientes hospitalares onde estabilidade é fundamental.

Levando em consideração o objetivo principal do projeto, que é ***construir um modelo preditivo confiável para ajudar a equipe médica a prever se uma pessoa tem obesidade — com base em fatores como idade, alimentação, atividades físicas, entre outros***, o modelo RANDOM FOREST foi escolhido pelos principais motivos:

* **Alta Precisão (acurácia de 99,19%):** Isso significa que, de cada 100 pessoas analisadas, o modelo acerta quase 99 casos — um desempenho excelente para um sistema de apoio à decisão médica.
* **Robustez:** Random Forest é um conjunto de várias “árvores de decisão”, cada uma votando sua opinião. A resposta final é uma média do “voto da floresta”, o que reduz erros individuais e torna o modelo mais estável e menos sensível a dados fora do padrão.
* **Evita Overfitting (erros por memorizar demais os dados):** Diferente de modelos simples que “decoram” os dados, o Random Forest consegue generalizar bem — ou seja, funciona bem com novos pacientes que ainda não estão no banco de dados.
* **Lida bem com diferentes tipos de dados:** O modelo aceita variáveis numéricas e categóricas (como “Fuma?”, “Usa transporte público?”, etc.) sem muito pré-processamento complexo. Isso torna a solução mais fácil de manter e atualizar no futuro.
* **Explicável:** Para aplicações médicas, explicabilidade é essencial. Random Forest permite mostrar:
 * Quais fatores mais influenciam a obesidade (ex: IMC, frequência de atividade física, consumo de alimentos calóricos)
 * O impacto de cada variável, ajudando os profissionais a entender “o porquê da previsão”
* **Segurança e Confiabilidade:** A menor variação nos resultados (±0.28%) mostra que o modelo é consistente. Ou seja, mesmo quando testamos com amostras diferentes da base de dados, ele mantém o bom desempenho.

**Avaliação final do Random Forest (classification report + matriz de confusão)**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Separar em treino e teste para avaliação final
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Treinar modelo final com Random Forest
final_model = RandomForestClassifier(random_state=42)
final_model.fit(X_train, y_train)

# Previsões
y_pred = final_model.predict(X_test)

# Relatório de Classificação
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()

# Matriz de Confusão
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoders['Obesity'].classes_, yticklabels=label_encoders['Obesity'].classes_)
plt.title("Matriz de Confusão - Random Forest")
plt.ylabel("Valor Real")
plt.xlabel("Previsão do Modelo")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

report_df
```
